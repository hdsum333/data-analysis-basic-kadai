```python
import sklearn
import pandas as pd

#Step1.
 #1. サンプルデータの取得
from sklearn.datasets import load_wine
dataset = load_wine()
df = pd.DataFrame(data=dataset.data, columns= dataset.feature_names)
df['category'] = dataset.target

 #2. サンプルデータの分割
  #説明変数（ワインの成分）を変数Xに、目的変数（ワインの種別）を変数yに代入
x = dataset.data
y = dataset.target

  #このXとyを、それぞれ学習データとテストデータに分割
  #model_selectionモジュールのtrain_test_split関数を実行することで、分割
from sklearn. model_selection import train_test_split

  #4つの要素をそれぞれ、変数X_train, X_test, y_train, y_testに代入
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=5)

#Step2.
 #3. 予測モデルのインスタンス化
  #予測モデルを選択し、クラスをインスタンス化、予測モデルとして決定木を選択
from sklearn.ensemble import RandomForestClassifier

  #RandomForestClassifierクラスをインスタンス化して、modelという変数に代入
model= RandomForestClassifier(random_state=3)

 #4. 予測モデルの学習
  #fitメソッドを実行することで、学習を実行
  #fitメソッドの引数に指定するデータは、説明変数X（ワインの成分）の学習データと目的変数y（ワインの種別）の学習データ
  #第1引数にX_trainを、第2引数にy_trainを指定
print(model.fit(x_train, y_train))

 #5. 予測モデルの評価
  #説明変数X（ワインの成分）のテストデータに基づいてワインの種別を予測した値と
  #目的変数y（ワインの種別）のテストデータがどれだけマッチしているかが評価のポイント
  #説明変数X（ワインの成分）のテストデータから、ワインの種別を予測
  #predictメソッドを利用することで予測値を出力
  #predictメソッドの引数には、説明変数XのテストデータであるX_testを指定
  #予測値を変数y_predに代入して出力
y_pred = model.predict(x_test)
print(y_pred)

#目的変数y（ワインの種別）のテストデータを確認
print(y_test)

#評価指標として正解率を利用
#metricsモジュールのaccuracy_score関数を実行することで、正解率が出力される
#y_testの54データのうち、y_predのデータがいくつヒットしたかという割合が正解率となる
#まずはaccuracy_score関数をインポート
from sklearn.metrics import accuracy_score

#インポートしたaccuracy_score関数を実行
#第1引数にテストデータであるy_testを、第2引数に予測値であるy_predをセット
print(accuracy_score(y_test, y_pred))
```
